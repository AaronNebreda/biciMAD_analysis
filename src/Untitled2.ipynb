```python
dat_barrios = bar_est[['CODDIS', 'NOMDIS', 'COD_BAR','NOMBRE','id', 'total_bases','n_salidas', 'n_entradas']]\
    .groupby(['CODDIS', 'NOMDIS', 'COD_BAR','NOMBRE'])\
    .agg(n_estaciones=('id', 'count'),
         N_bases = ('total_bases', 'sum'),
        n_salidas =('n_salidas','sum'),  
         n_entradas =('n_entradas','sum'),
        ).reset_index()






```

```python
def ratio(x,y):
    if x > y:
        return 1 - (1 - ((y + 0.0000001)/x))
    elif x < y:
        return 1 + (1 - ((x + 0.0000001)/y))
    else:
        return 1
```

```python

dat_barrios['n_salidas_AVG'] = dat_barrios['n_salidas']/365
dat_barrios['n_entradas_AVG'] = dat_barrios['n_entradas']/365
dat_barrios['ratio1'] = dat_barrios.apply(lambda x : ratio(x['n_salidas'], x['n_entradas']), axis=1)
dat_barrios['mov_TOT'] =  dat_barrios['n_entradas'] +  dat_barrios['n_salidas']
dat_barrios['mov_TOT_AVG'] =  dat_barrios['mov_TOT']/ 365
```
